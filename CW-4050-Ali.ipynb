<h2>Data Processing</h2>

Used Feature Importance to reduce dimentionality of the dataset,
And used Extra Tree Classifier to compute out important features.

<b>Extra tree Classifier</b> 
The following algorithm is used to compute important features from high dimentional data, used to do feature selection whereas
the algorithm selects a split point at random.The random selection of split points makes the decision trees in the ensemble less correlated, although this increases the variance of the algorithm.

```python
#imported ExtraTreeClassifier from sklearn.
from sklearn.ensemble import ExtraTreesClassifier

#defined model and fitted to the dataset.
model = ExtraTreesClassifier()
model.fit(X,y.values.ravel())
```

```python
#use inbuilt class feature_importances of tree based classifiers
print(model.feature_importances_) 

#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)

#plotted for 20 imortant features that are impactful.
feat_importances.nlargest(20).plot(kind='bar',figsize=(15,10))

#used dark them from seaborn
sns.set_palette("dark")

plt.grid()
plt.show()
```

```python
#Manualy selected important features computed by ETC and assign to Variable X.
X= X[['CCL5','PLAAT4','PPP2R2B','TMEM218','RHOF','MCUB','SKAP1','IGSF9B','DENND1C','IGFBP3','MAF','KLRG1','PRR5L','SORBS3','MFSD5','EXOSC1','NCR3','GGT7','BANF1','PLXDC1']]

X
```

<h2>Training and testing</h2>

Used train test split technique to keep model simple and to remove complexity, because model already used Extra tree classifier to reduce high dimensional data.    

```python
#imported train test split from sklearn 
from sklearn.model_selection import train_test_split

#splits our dataset into train and test data.
#Training data is composed by 67% of data, the remaining 33% of data goes to the test set(kept unseen).
#set random state to 1.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1, shuffle=True)
```

```python
#checked for the shape after applyig train test split.
X_train.shape
```

```python
y_train.shape
```

<h2>Pipeline using Logistic Regression</h2>
Proposed pipeline is stated as below:

Standardisation
Logistic regression

Used Logistic Regression model from sklearn.linear_model.

The C can be used to alter the “sensitivity” of the algorithm. I would start off with a default value of 1.
May check with different C value offering the best bias-variance trade-off. 

However, since I have applied Extra Tree Classifier to reduce dimentionality, I do not expect the model to have very high variance, therefore it may not need strong regularization and tuning.

```python
 # used Library to standardise data.
from sklearn.preprocessing import StandardScaler

# used Library for Logistic Regression
from sklearn.linear_model import LogisticRegression 

# used Library to create a pipeline
from sklearn.pipeline import Pipeline 

# Building a pipeline
pipe = Pipeline([('scaler', StandardScaler()), # scaling features
('logreg', LogisticRegression(C=1, random_state=0))]) # logistic regression

# Implementing the pipe
pipe.fit(X_train, y_train.Severity)
    
# Use the trained model to create predictions on train and test data
y_train_pred = pipe.predict(X_train)
y_test_pred = pipe.predict(X_test)
```

USED ACCURACY TEST METRIC TO MEASURE ACCURACY OF THE MODEL. 

<h2>Accuracy</h2>

Checked for the model performance on train test data.

Created two seperate DataFrames for train and test data respectively.

Datasets contains actual and fitted values. 

```python
train = pd.DataFrame({"actual": y_train.Severity.values,
                      "fitted": y_train_pred})
train.head()
```

```python
#testing model is generalisable on unseen data
test = pd.DataFrame({"actual": y_test.Severity.values,
                     "fitted": y_test_pred})
test.head()
```

Used accuracy_score from sklearn.metrics to compute out the accuracy of proposed model.

```python
from sklearn.metrics import accuracy_score

print("Accuracy on train:", accuracy_score(train.actual, train.fitted))
print("Accuracy on test:", accuracy_score(test.actual, test.fitted))
```

The accuracy of our model is good enough; furthermore accuracy on train and test data are quite close.

<h2>Precison, Recall, F-Score</h2>

Imported precision_recall_fscore_support from sklearn.metrics which is used to compute the precision, recall and f-score of predictions. 


<b>Precision:</b> Quantifies the number of positive class predictions that actually belong to the positive class.

<b>Recall:</b> Quantifies the number of positive class predictions made out of all positive examples in the dataset.

<b>F-Score:</b> Provides a single score that balances both the concerns of precision and recall in one number.

```python
#used library to computes precision, recall and f-score for each class of test data
from sklearn.metrics import precision_recall_fscore_support

#average is set to none yet
precision_recall_fscore_support(test.actual, test.fitted, average=None)
```

Each class has the output precision of 0.80952381 means score results are pretty good.
